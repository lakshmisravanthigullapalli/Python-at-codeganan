# dictionary methods
-get(key,default_value:optional):it returns the value if key is exists in dictionary
otherwise it returns default value

```python
details={"name":"sravanthi",1 :100,"course":"python",(1,2,3,4):57,"age":20}
print(details['name'])
#$print(details['marks'])
print(details.get('name'))
print(details.get('marks',"marks not found"))
```

-update(new_dictionary):This method updates the values if new_dictionary keys already exist in dictionary ,otherwise it add new items to dictionary

```python
details={"age":25,"name":"sravanthi",1 :100,"course":"python","Batch":57}
details.update({'Batch':58,'course':'java',"location":"hyd"})
print(details)
```

-keys():This method returns the list dictionary keys
-values():This method returns the list dictionary values
-items():This method returns the list of tuple of key-value pairs 

```python
details={'age': 25, 'name': 'sravanthi', 1: 100, 'course': 'java', 'Batch': 58, 'location': 'hyd'}

print(details.keys())
keys=details.keys()
print(type(keys))
print(details.values())
print(details.items())
```

-popitem():It return and removes the last item as tuple
-pop key():this method removes the specified key item and return the specified keys values otherwise keyerror

```python
details={'age': 25, 'name': 'sravanthi', 1: 100, 'course': 'java', 'Batch': 58, 'location': 'hyd'}
item=details.popitem()
print(item)
print(details)
val=details.pop('name')
print(val)
print(details)
```
