### Set Methods

```python
# Element adding method
# add(ele):It adds the element to set if element not in set,otherwise nothing
s={1,2,3,4,5,4,3}
print(s)
s.add(100)
s.add(5)
s.add("code")
print(s)
```

```python
# element removing methods 
#remove (item):It removes the item if it exists in set otherwise "key error"
#discard (item):It removes the item if it exists in set otherwise nothing
s={1, 2, 3, 4, 5}
s.remove(5)
print(s)
s.remove(100)#keyerror
s.discard(3)
print(s)
s.discard(100)# it will not stop the program not rise error prefer to use
print(s)
```

```python
# set operations
#union (set):It returns all the elements from all sets
# any method returning value assign to variable
s1={1,2,3,7}
s2={4,5,6,7}
s3={1,2,8,9}
union=s1.union(s2,s3)
print(union)
res=s1 |s2|s3
print(res)
```

```python
# set operations
#intersection:It returns all the common elements from all sets
# any method returning value assign to variable
s1={1,2,3,7}
s2={4,5,6,7}
s3={1,2,8,7}
intersection=s1.intersection(s2,s3)
print(intersection)
res=s1&s2&s3
print(res)
```

```python
s1={1,2,3,7}
s2={4,5,6,7}
s3={1,2,8,9}
intersection=s1.intersection(s2,s3)#no common elements
print(intersection)
```

```python
# # set operations
#difference:It returns the  elements that are not in another sets from all sets
# any method returning value assign to variable
s1={1,2,3,7}
s2={4,5,6,7}
s3={1,2,8,7}
intersection=s1.difference(s2,s3)
print(intersection)
res=s1-s2-s3
print(res)
res=(s1|s2)-s3
print(res)
```

```python
# set operations
#symmetric difference:It returns the non common elements from all sets
# any method returning value assign to variable
s1={1,2,3,7}
s2={4,5,6,7}
s3={1,2,8,7}
sy_diff= s1.symmetric_difference(s2)#one argument is allowed
print(sy_diff)
res=s1^s2^s3#method we can give
print(res)
```
