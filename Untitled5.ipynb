```python
def printnumbers(n):
    for i in range(1,n):
        print(i,end=" ")
    print(n)    
m=int(input())
printnumbers(m)

```

```python
n=int(input())
res=""
while n>0:
    rem=n%2
    res=str(rem)+res
    n=n// 2
print(res)
```

```python
#reverse a string without using slicing
string=input()
rev=""
for char in string:
    rev=char+rev
print(rev)
```
