```python
#find bigger and smaller characters from s1 and s2
#compare each character in both strings and if s1[i]>s2[i] to bigger and s2[i] to smaller
# if both same leave it
s1 ,s2 = map(str, input().upper().split())
n = len(s1)
big = small = ""
for i in range(n):
    if s1[i] != s2[i]:
        if s1[i] > s2[i]:
            big += s1[i]
            small += s2[i]
        else:
            big += s2[i]
            small += s1[i]
print(big)
print(small)
```

```python
#map(fun,iterable):IT applies the fuctionality on each element in iterable
#filter(fun,iterable):It filters the itables,it returns the value which is not none
#find the square of list of integers
def square(x):
    return x**2
lst=[1,2,3,4,5]
print(list(map(square,lst)))
        
```

```python
#filter the even numbers from list of integers
def even(x):
    if x%2==0:
        return x
lst=[1,2,3,4,5,6,7]
print(list(filter(even,lst)))#filter function
print(list(map(even,lst)))#map function
```

```python
def square(t):
    x,y=t
    return x+y
lst=[(1,2),(3,4)]
print(list(map(square,lst)))
        
```

```python
# find the target element position in list
arr=[10,20,30,40,50]
target=40
for ind,val in enumerate(arr):
    if target==val:
        print(ind)
        break
else:
    print(-1)
```

```python
#any(iterable):IT RETURNS true when any one element in iterable boolean value is true
arr=[1,2,3,0,0,0]
print(any(arr))
print(any([0,0,0,0]))
print(all([1,2,3,4]))
print(all([True,False,True]))
print(all([False,False,False]))
#all(iterable):IT RETURNS true when all iterable elements boolean value is true
```
