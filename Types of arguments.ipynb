```python
#Types of arguments
#1.positional Arguments:here arguments are assigns to parameters as per the sequential order
def sub(x:int,y:int)->int:
    z=x-y
    return z
a,b=20,10
print(sub(a,b))
print(sub(b,a))
```

```python
#keyword arguments:arguments are explicitly passed with parameters names
def sub(x:int,y:int)->int:
    z=x-y
    return z
a,b=20,10
print(sub(x=a,y=b))
print(sub(y=b,x=a))
```

```python
#3.default arguments:the default arguments are optional arguments,if u didnt
#pass value for default argument, inthat case it assigns default value otherwise,
#                                                         it assigns new argument value
def add(x:int,y:int,z:int=0,w:int=0)->int:
    e=x+y+z+w
    return e
a,b,c,d=map(int,input().split())
#print(add(a,b,c,d))
print(add(a,b))
print(add(a,b,c))
print(add(a,b,c,d))
```

```python
# variable length positional arguments
def add(x,y=0,*args):
    print(type(args))
    print(args)
    s=0
    for val in args:
        s+= val
    return s
        
print(add(10,20,1,2,3,4,5))
```

```python
# variable length positional arguments
def add(x,y=0,*args):
    print(type(args))
    print(x,y,args)
    s=0
    for val in args:
        s+= val
    return s
        
print(add(10,20,1,2,3,4,5))
```

```python
# variable length keywords arguments
def add(x,y=0,*args,**kwargs):
    print(type(kwargs))
    print(kwargs)
    print(kwargs.values())
    print(kwargs.keys())
print(add(1,2,3,4,5,6,name="sravanthi",age=20,course="pfs"))
```

```python
#pass by reference
def add_ele(lst,ele):#pointing to same memory allocation
    lst.append(ele)
arr =[1,2,3,4]
print("before execution arr value:",arr)
add_ele(arr,100)
print("after function arr value:",arr)
```
