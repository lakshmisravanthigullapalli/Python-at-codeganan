```python
#if else
num=int(input())
if num%3==0:
    print(f"{num} is divisible by 3")
else:
    print(f"{num} is divisible by not 3")
    
```

```python
#check even or odd
num=int(input())
if num%2==0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
```

```python
#check even or odd
num=int(input())
if num & 1==0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
```

```python
#5th question
#dictionary dont take from the user for the products list
products ={'pen':5,'book':20,"pencil":3}
item=input().lower()
quantity=int(input())
if item in products:
    total_amount=products.get(item)*quantity
    print(f"Total amount of{quantity}{item}:{total_amount}")
else:
    print(f"items not found")
```

```python
#check weather the number is positive or negative or zero
num=int(input())
if num>0:
    print(f"{num}is positive")
elif num<0:
     print(f"{num}is negative")
else:
     print(f"{num}is zero")
```

```python
#divisible by 3-fizz
#divisible by5-buzz
#both-fizz buzz
num=int(input())
if num%3==0:
    print(f"{num} is fizz")
elif num%5==0:
    print(f"{num} is buzz") 
elif num%3==0 and num%5==0:
    print("fizz buzz")
else:
    print("invalid")  
    
```

```python
#divisible by 3-fizz
#divisible by5-buzz
#both-fizz buzz
num=int(input())#here conditions have prioity
if num%3==0 and num%5==0:
    print("fizz buzz")
elif num%3==0:    
    print(f"{num} is fizz")
elif num%5==0:
    print(f"{num} is buzz") 

else:
    print("invalid")  
    
```
