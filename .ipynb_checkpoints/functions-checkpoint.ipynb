```python
# types of functions based on parameters and return value
#4.with parameters and with return
#fun definition block
def addition(x:int,y:int)->int:# rayaka poyina ntg
    c=x+y
    return c
    #function calling block
a,b=10,20
res=addition(a,b)
print(res)
```

```python
#4.with parameters and with0ut return
#fun definition block
#by default fun returns none value,when fun not returning any value
def addition(x:int,y:int)->None:# rayaka poyina ntg
    c=x+y
    #return 
    #print("hi")
    #function calling block
a,b=10,20
res=addition(a,b)
print(res)#fun  return lekapote none print ayidi
```

```python
#4.with parameters and with0ut return
#fun definition block
#by default fun returns none value,when fun not returning any value
def addition(x:int,y:int)->int:# rayaka poyina ntg
    c=x+y
    #return 
    print("this is from fun definition block",c)
    #function calling block
a,b=10,20
res=addition(a,b)
print("this is from fun calling",res)#fun run ayidi return lekapoyina none print ayidi
```

```python
#2. without parameters and with return
def addition()->int:
    x= 10
    y=20
    z=x+y
    return z
    #fun calling
res=addition()
print("this is from function calling block:",res)
```

```python
#2. without parameters and without return
def addition()->int:
    x= 10
    y=20
    z=x+y
    print("this is from function definition block:",z)
    #fun calling
res=addition()
print("this is from function calling block:",res)
```

```python
#even or odd
def evenodd(x:int):
    if x%2==0:
       return"even"
    else:
        return"odd"
a=int(input())
res=evenodd(a)
print(res)
```

```python
#prime numbers using function
def prime(n:int):
    if n<2:
       return"not prime"
    for i in range(2,n//2+1):
        if n%i==0:
           return "not prime"
    return "prime"        
num=int(input())
print(prime(num))
```
