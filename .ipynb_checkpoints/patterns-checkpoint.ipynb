```python
n=5

for i in range(n):
    for j in range(n):
         if i>=j:#for shape 
             print(chr(65+j),end=" \t")
            
         else:
             print(" ",end="\t")
    print()         
```

```python
n=5
k=n*(n+1)//2
for i in range(n):
    for j in range(n):
         if i>=j:#for shape 
             print(k,end=" \t")
             k-=1
         else:
             print(" ",end="\t")
    print()         
```

```python
n=8
k=0
for i in range(n):
    for j in range(n):
         if i>=j:#for shape 
             print(chr(65+(k%26)),end=" \t")
             k+=1
         else:
             print(" ",end="\t")
    print()         
```

```python
n=5
k=0
for i in range(n):
    for j in range(n):
         if i>=j:#for shape 
             print(chr(65+k),end=" \t")
             k+=1
         if k==26:
              print(chr(65+k),end=" \t")
         else:
             print(" ",end="\t")
    print()         
```
