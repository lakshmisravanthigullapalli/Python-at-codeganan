```python
#scope of variables
#1.local scope
#2.Enclosing scope
#3.global scope
#4.builtin scope
x=10#global scope variable
def outer():
    y=20 #enclosing variable
    def inner():
        z=30 #local scope variable
        print(x,y,z) #builtin scope variable
    inner()
outer()    
```

```python
#using global scope variable in local scope
x=100
def outer():
    y=x+1
    print("local scope variable value:",y)
outer()
print("Global scope variable value:",x)
    
    
```

```python
#updating the global scope variable value in local scope
#cannot update global scope variable in local scope directly
x=10
def outer():
    x=x+1
    print("local scope variable:",x)
outer()
print("Global scope variable value:",x)
    
        
```

```python
#updating the global scope variable value in local scope by using "global" keyword
x=100
def outer():
    global x # update in entire globally
    x=x+1
    print("local scope variable:",x)
outer()
print("Global scope variable value:",x)
```

```python
#updating the enclosing scope variable value in local scope
x=10 #global scope variable
def outer():
    y=20 #enclosing variable
    def inner():
        z=30 #local variable
        y=y+z
        print(y)
    inner()
outer()    
```

```python
x=10 #global scope variable
def outer():
    y=20 #enclosing variable
    def inner():
        nonlocal y
        z=30 #local variable
        y=y+z
        print(z,x,y)
    inner()
    print("enclosing value:",y)
outer()    
```

```python
x=10 #global scope accessing in local scope using same variable not update in global
def outer():
    x=10
    x=x+2
    print("local scope variable x value:",x)
outer()
print("global scope variable x value:",x)
```
