### list methods 
 -Item adding methods
-append(item):it adds item at end of list
-extend(iterator):it adds the all iterator elements at the end of list
insert(pops,item):it adds item at specifie dposition

```python
lst=[10,20,30,40]
print(lst.append(100))
print(lst)
lst.extend([1,2,3])
print(lst)
lst.insert(0,1000)
print(lst)
lst.append([100,200])
print(lst)
print(lst)
```

### item removing methods
    - remove(item): It removes the  first occurence item from the list it is exists in list otherwise
    value error
        -pop(pos): it remoove and return the specified position item ifposition is passed
                    otherwise it remove and returns the last item from list

```python
lst=[10,20,30,20,40,50,20]
print(lst.remove(20))
print(lst)
lst.remove(100)
print(lst)
```

```python
lst =[10,20,30,20,40,50,20]
val=lst.pop()
print(val)
print(lst)
val2=lst.pop(0)#it removes the 0th position
print(val2)
print(lst)
lst.clear()
print(lst)
```

```python
# index and count()
#index(item): if item is present in list it returns occurence the index otherwise value error
#count(item):it return the count of total occurences of item
lst =[10,20,30,20,40,50,20]
ind =lst.index(20)
print(ind)
print(lst.count(20))
```

```python
#sorting items
#sort(reverse=false):by default it sorts the origin list in into ascending order
#                       if reverser is true it sorts in decending order
lst=[1,4,2,5,-5,3,-45,-4,3]
print(lst.sort())
print(lst)
print(lst.sort(reverse=True))
print(lst)

     
              
```
