```python
n1,n2,n3=map(int,input().split())
if n1%2==0:
    print(f'{n1} is even')
else:
    print(f'{n1} is odd')
if n2%2==0:
    print(f'{n2} is even')
else:
    print(f'{n2} is odd')   
if n3%2==0:
    print(f'{n3} is even')
else:
    print(f'{n3} is odd')    
```

```python
lst=[1,2,3]
for num in lst:
    if num%2==0:
        print(f'{num} is even')
    else:
        print(f'{num} is odd')
        
```

```python
# check which one is even or odd
lst=list(map(int,input().split()))
for num in lst:
    if num%2==0:
        print( num**2)
    else:
        print(num+1)
        
```

```python
lst=list(map(int,input().split()))
for num in lst:
    if  num%3==0 and num%5==0:
        print("fizz buzz")
    elif num%3==0:
        print("fizz")
    elif num%5==0:
        print("buzz")
```

```python
lst=list(map(int,input().split()))
res=[]
for num in lst:
    if  num%3==0 and num%5==0:
        res.append("fizz buzz")
    elif num%3==0:
        res.append("fizz")
    elif num%5==0:
        res.append("buzz")
print(res)
```

```python
s=input()
v=0
c=0
for char in s:
    if char in 'aeiou':
        v+=1
    else:
        c+=1
print(f'vowels{v} ,consonant{c}')
```

```python
s=input().lower()
v=0
c=0
for char in s:
  if char.isalpha():
     if char in 'aeiou':
        v+=1
     else:
        c+=1
print(f'vowels{v} ,consonant{c}')
```
