```python
#Direct Recursion
# print 1 to n using recursion
def fun(x):
    if x==0:
        return 0
    fun(x-1)
    print(x)
        
fun(3)
```

```python
def fun(x):
    if x==0:
        return 0
    
    print(x)
    fun(x-1)   
print(fun(3))
```

```python
#factorial
def factorial(x):
    if x==0 or x==1:
        return 1
    else:
        return x*factorial(x-1)
print(factorial(5))     
```

```python
# sum of digits in a number
#n=456 o/p:15
def sumofdigits(n):
    if n==0:
        return 0 
    return n% 10+sumofdigits(n//10)
print(sumofdigits(456))
print(sumofdigits(15346))
```

```python
# count digits in a number
#n=456 o/p:15
def sumofdigits(n):
    if n==0:
        return 0 
    return 1+sumofdigits(n//10)
print(sumofdigits(456))
print(sumofdigits(15346))
```

- # lamda functions or Anonymous functions:

```python
#find the square of a number using lamda
s= lambda x:x**2
print(s(10))
```

```python
#find the squares of list of integers using map and lambda
arr=[1,2,3,4,5]
print(list(map(lambda x: x**2,arr)))
```

```python
arr=[1,2,3,4,5]
print(list(filter(lambda x: x%2==0,arr)))
```

```python
s=lambda x:"even" if x%2==0 else "odd"
print(s(3))
```

```python
#arr=[1,2,3,4,5]
#1-odd
 #2-even
arr=[1,2,3,4,5]
print(list(map(lambda x:f"{x}-even" if x%2==0 else f"{x}-odd",arr)))     
```

```python
#find the frequency of each character
string=input()
count={}
for char in string:
    if char in count:
        count[char]+=1
    else:
        count[char]=1
print(count)
```

```python
string=input()
count={}
for char in string:
    count[char]=count.get(char,0)+1
print(count)
```
