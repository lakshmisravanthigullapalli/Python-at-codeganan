```python
#generate 1-n numbers using list comprehension
n=int(input())
res=[x for x in range(1,n+1)]
print(res)
```

```python
#first n even numbers using list comprehension
n=int(input())
res=[x*2 for x in range(1,n) ]
print(res)
```

```python
#find the first n cubic numbers
n=int(input())
res=[x**3 for x in range(1,n+1)]
print(res)
```

```python
#find odd numbers between n1 t0 n2
n1,n2=map(int,input().split())
res=[x for x in range(n1,n2+1)if x%2==1]
print(res)
```

```python
#in a given arr of integers perform following operations
#if even-find square of a number
#if odd-find squareroot value(upto 2 decimal places)
arr=list(map(int,input().split()))
res=[x**2 if x%2==0  else x**0.5 for x in arr]
print(res)
```

```python
arr=list(map(int,input().split()))
res=[x**2 if x%2==0  else round(x**0.5,2) for x in arr]
print(res)
```

```python
#2 dices combination values
res=[(x,y)for x in range(1,7) for y in range(1,7)]
print(res)
```

```python
# return the dice combination if sum of two face value is divisible by 3
res=[(x,y) for x in range(1,7) for y in range(1,7) if(x+y)%3==0]
print(res)
```

```python
res=[(x,y)for x in range(1,7) for y in range(1,7) if x==4 or y==4]
print(res)
```

```python
# count the number of vowels exists in string
string=input()
res=[char for char in string.lower() if char in"aeiou"]
print(res)
print(len(res))
```

```python
# count the number of vowels existing in the string
s = input().lower()
res = [x for x in s if x in "aeiou"]
print(res)
print(len(res))
```

```python
#dictionary comprehension
#1 to n numbers
n=int(input())
res={x:x**2 for x in range(1,n+1)}
print(res)
print(type(res))
```

```python
#set comprehension
#1 to n numbers
n = int(input())
res = {x for x in range(1,n+1)}
print(type(res))
print(res)
```
