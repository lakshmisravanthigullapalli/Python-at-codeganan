```python
# reading integer from user
#input formatting
num=int(input("enter a number:"))
print(type(num))
print(num)
```

```python
#reading float value from user
val=float(input())
print(type(val))
print(val)
```

```python
# complex reading from user
val=complex(int(input()),int(input()))
print(type(val))
print(val)
```

```python
# boolean reading
val=bool(input())
print(type(val))
print(val)
```

```python
#boolean reading as numbers-it will take only numbers notcharacters first string to int next int to bool
val=bool(int(input()))
print(type(val))
print(val)
```

```python
#reading string value from the user
s=input()
print(type(s))
print(s)
```

```python
# readding string to list
s=list(input())
print(type(s))
print(s)
```

```python
# reading list of intger from user

lst=list(map(int,input().split()))
print(lst)
```

```python
# reading list of intger from user

lst=list(map(int,input().split(',')))
print(lst)
```

```python
# reading tuple of integers from user

lst=tuple(map(int,input().split()))
print(lst)
```

```python
#eval():this function evaluate the string type expressio n
val=eval(input())
print(type(val))
print(val)
```

```python
print("3+5")
print(eval("3+5"))
```

```python
#1.string formatting using commas
name="sravanthi"
age=20
print("my name is ",name," and i am ",age," years old")
```

```python

name="sravanthi"
age=20
print("my name is "+name+" and i am "+str(age)+" years old")
```

```python
#access specifiers
#string-%s
#integer-%d
#float-%f
name="sravanthi"
age=20
print("my name is %s and i am %d years old"%(name,age))
```

```python
name="sravanthi"
age=20
print("my name is %s and i am %d years old"%(age,name))
```

```python
#3 dot format
name="sravanthi"
age=20
print("my name is {} and i am {} years old".format(name,age))
print("my name is {} and i am {} years old".format(age,name))
```

```python
#3 f- format
name="sravanthi"
age=20
print(f"my name is {name} and i am {age} years old")
print(f"my name is {age} and i am {name} years old")
```
