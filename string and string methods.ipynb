## string methods
- Case conversion methods
      -upper():it converts entire string into upper case
      -lower():it converts entiiire string into lower case
       -title():it converts the first letter of every word into upper and remain lower
       -swapcase():it convert upper case letter into lowercase and vice versa
       - capitalize():it converts first character into upper case and remaining into lower
  

```python
string = "The Sky is blue and Ocean is BLUE@123"
print(string.upper())
print(string.lower())
print(string.title())
print(string.swapcase())
print(string.capitalize())
```

- remove white Space or Pattern at ends of string
      -strip(): By default it removes the white space at both ends of string
      -lstrip(): it removes the white space at left side of string
       -rstrip(): it removes the white space at right side of string

```python
s= "      codegnan    "
print(s)
print(s.strip())
print(s.rstrip())
print(s.lstrip())
```

```python
s= "@@@@@codeg@@nan@@@@"
print(s)
print(s.strip("@"))
print(s.rstrip("@"))
print(s.lstrip("@"))
```

```python
# split(): it splits the entire string into list of substrings based on specified seperator
# by default it splits at space
string = "The Sky is blue and Ocean is Blue"
print(string.split())
print(string.split("is"))
print(string.split("e"))
```

```python
# 'sep'.join(iterable): it joins the all iterable elements with seperator
words=['The', 'Sky','is', 'blue', 'and', 'Ocean', 'is', 'Blue']
print("".join(words))
print(" ".join(words))
print("-".join(words))
print("/".join(words))

```

## substring index finding
- index(substring): if substring is existing in the string ,it returns the first occuerence of the substring,otherwisw value error
- find(substring): if substring is existing in the string ,it returns the first occuerence of the substring,otherwisw -1

```python
string="The Sky is blue and Ocean is Blue"
print(string.index("is"))
print(string.index('z'))
```

```python
string="The Sky is blue and Ocean is Blue"
print(string.find("is"))
print(string.find('z'))
print(string.find('is',10))# it starts checking from the 10th position
print(string.find('e',10))
print(string.find('e',-5))

```

```python
# count (substing): it returns the total non-overlapping occucences of substing
s="ABCBCBCbbaabb"
print(s.count("B"))
print(s.count("BCB"))
lower_string= s.lower()
print(lower_string)
print(lower_string.count("b"))
print(s.upper().count('B'))
print(s.count('z'))
s="ABCBCBCBbbaabb"
print(s.count("BCB"))
```

```python
# replace (old_string,new_string):it replaces the old substring with the new substring
string="The Sky is blue and Ocean is Blue"
print(string.replace("blue","red"))
print(string.replace("e","E"))
print(string.replace("e","E",2))# it replaces the first two occurences
```

### String checking methods 
- string checking the methods returns the boolean value as result
- isupper():It return true ,if entire string is in upper case
-  islowerr():It return true ,if entire string is in lower case
-   isalpha():It return true ,if entire string contains alphabets
-    isdigit():It return true ,if entire string contains digits
-     isalnum():It return true ,if entire string contains either alphabetsor numbers
- startswith (substring):it returns true, if string starts with substring
- endswith (substring):it returns true, if string ends with substring
- 

```python
print("CODE".isupper())
print("CODE".islower())
print("code".islower())
print("1234".isdigit())
print("1234abc".isdigit())
print("code".isalpha())
print("code1234".isdigit())
print("code".isdigit())
print("1234".isalnum())#checks eithe alpha or num
print("code1234".isalnum())
print("code".startswith("c"))
```
