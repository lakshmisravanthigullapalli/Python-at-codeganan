```python
lst=list(map(int,input().split()))
for num in lst:
    if  num%3==0 and num%5==0:
        print("fizz buzz")
    elif num%3==0:
        print("fizz")
    elif num%5==0:
        print("buzz")
```

```python
#prime number
n=int(input())
for i in range(2,n//2+1):
    if n%i==0:
        print(f"{n} not prime")
        break
else:
    print(f"{n} is prime")
        
```

```python
#prime number
m=int(input())
for n in range(2,m):
    for i in range(2,n//2+1):
        if n%i==0:
            print(f"{n} not prime")
            break
else:
    print(f"{n} is prime")
        
```
